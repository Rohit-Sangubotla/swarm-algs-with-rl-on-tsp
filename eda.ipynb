```python
import pandas as pd
```

```python
algs = ['aco','antq','pso']
cities_nums = [5,25,50,100]

for cities_num in cities_nums:
    series = [0,1,2]
    for i in range(len(algs)):
        alg = algs[i]
        df = pd.read_csv(f'results/{alg}_results_{cities_num}.csv')
        series[i] = df['best_tour_length']
        series[i].rename(f'{alg}', inplace=True)
    alg_df = pd.concat(series, axis=1)
    print(f'{cities_num} cities')
    print(alg_df)
    alg_df.to_csv(f'results/{cities_num}_cities.csv', index=False)
```

```python
avgs = []

for cities_num in cities_nums:
    df = pd.read_csv(f'results/{cities_num}_cities.csv')
    avgs.append(df.mean())

avgs = pd.concat(avgs, axis=1)
avgs = avgs.rename(columns={0: '5 cities', 1: '25 cities', 2: '50 cities', 3: '100 cities'})
avgs.to_csv('results/avgs.csv', index=False)
```
